```python
#Load and Inspect the Dataset
import pandas as pd

# Load dataset
df = pd.read_csv("t20_worldcup_team_features_large_1000.csv")

# Display basic information
print("Dataset Info:")
print(df.info())

# Show the first 5 rows
print("\nSample Data:")
print(df.head())
```

```python
#Preprocess the Data
from sklearn.preprocessing import LabelEncoder

# Encode team names as numeric values
le = LabelEncoder()
df['Team_encoded'] = le.fit_transform(df['Team'])

# Split data
df_train = df[df['Year'] < 2025].copy()
df_2025 = df[df['Year'] == 2025].copy()

# Features to use for training and prediction
features = [
    'Team_encoded', 'ICC_Rank', 'Batting_Rating', 'Bowling_Rating',
    'Win_Rate_Last5', 'Group_Stage_Wins', 'Avg_Strike_Rate', 'Avg_Economy_Rate'
]

# Prepare training and test data
X_train = df_train[features]
y_train = df_train['Tournament_Winner']
X_2025 = df_2025[features]
```

```python
#Train a Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict probabilities for 2025 teams (probability of winning)
rf_probs_2025 = rf_model.predict_proba(X_2025)[:, 1]  # Probability of class 1 (winner)

# Add predictions to 2025 dataframe
df_2025['RF_Win_Probability'] = rf_probs_2025

# Show sorted predictions
rf_results = df_2025[['Team', 'RF_Win_Probability']].sort_values(by='RF_Win_Probability', ascending=False).reset_index(drop=True)

print("üèÜ Random Forest Predicted Win Probabilities (2025):")
print(rf_results)
```

```python
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Step 1: Normalize input features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_2025_scaled = scaler.transform(X_2025)

# Step 2: Convert target to categorical (0 -> [1, 0], 1 -> [0, 1])
y_train_dl = to_categorical(y_train)

# Step 3: Build the neural network
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')  # 2 classes: win or not
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X_train_scaled, y_train_dl, epochs=50, batch_size=16, verbose=0)

# Predict win probability for 2025 teams
dl_probs_2025 = model.predict(X_2025_scaled)[:, 1]  # Probability of being winner
df_2025['DL_Win_Probability'] = dl_probs_2025
```

```python
#Train a Deep Learning Model (Keras)
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Step 1: Normalize input features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_2025_scaled = scaler.transform(X_2025)

# Step 2: Convert target to categorical (0 -> [1, 0], 1 -> [0, 1])
y_train_dl = to_categorical(y_train)

# Step 3: Build the neural network
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')  # 2 classes: win or not
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X_train_scaled, y_train_dl, epochs=50, batch_size=16, verbose=0)

# Predict win probability for 2025 teams
dl_probs_2025 = model.predict(X_2025_scaled)[:, 1]  # Probability of being winner
df_2025['DL_Win_Probability'] = dl_probs_2025
```

```python
#Display and Compare Predictions
# Select relevant columns for comparison
comparison = df_2025[['Team', 'RF_Win_Probability', 'DL_Win_Probability']].copy()

# Sort by DL predicted win probability
comparison_sorted = comparison.sort_values(by='DL_Win_Probability', ascending=False).reset_index(drop=True)

# Display top 5 predictions
print("üèè T20 World Cup 2025 Winner Prediction:")
print(comparison_sorted.head(5))
```

```python
#Explainable AI (SHAP Values for DL or RF)
#use SHAP (SHapley Additive exPlanations) to show why a team is predicted to win.
import shap
import matplotlib.pyplot as plt
import numpy as np

# Use the correct variable name for the trained Random Forest model
explainer = shap.TreeExplainer(rf_model)
# Use the correct variable name for the training features
shap_values = explainer.shap_values(X_train)

# Calculate the mean absolute SHAP value for each feature across all classes
# The shape of shap_values is (n_samples, n_features, n_classes)
# We want to average across samples (axis 0) and classes (axis 2)
# Or more commonly for overall importance, average absolute SHAP values per feature across samples and classes
# Let's calculate the mean absolute SHAP value for each feature
mean_abs_shap_values = np.mean(np.abs(shap_values), axis=(0, 2))

# Create a DataFrame for plotting
feature_names = X_train.columns
shap_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Mean Absolute SHAP Value': mean_abs_shap_values
})

# Sort for plotting
shap_importance_df = shap_importance_df.sort_values(by='Mean Absolute SHAP Value', ascending=False)

# Plot SHAP summary bar plot using matplotlib/seaborn
plt.figure(figsize=(8, 5))
# Using seaborn for better visualization
import seaborn as sns
sns.barplot(data=shap_importance_df, x='Mean Absolute SHAP Value', y='Feature', palette='viridis')
plt.title('Feature Importance (Mean Absolute SHAP Value) - Random Forest')
plt.tight_layout()
plt.show()

# If you specifically want to plot for Class 1 (Winner), you can still do:
# Ensure X_train is a numpy array if shap_values is a numpy array from the explainer
# shap.summary_plot(shap_values[1], X_train.values, plot_type="bar", feature_names=feature_names)
# Note: Passing X_train.values (numpy array) explicitly might sometimes resolve subtle dataframe vs array issues with shap.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # Import LogisticRegression
from sklearn.ensemble import VotingClassifier # Import VotingClassifier

# Split the original training data (Year < 2025) into training and testing sets
X_train_split, X_test, y_train_split, y_test = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train # Using stratify is good for classification
)

# Now, use X_train_split and y_train_split for training your models (RF, LogReg, etc.)
# And use X_test and y_test for evaluation.

# Re-train the RandomForestClassifier using the split training data
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_split, y_train_split)

# Re-train the LogisticRegression model using the split training data
logreg = LogisticRegression()
logreg.fit(X_train_split, y_train_split)

# Re-define and train the voting classifier
voting = VotingClassifier(estimators=[
    ('rf', rf_model),
    ('lr', logreg)
], voting='soft')

voting.fit(X_train_split, y_train_split)

# Now you can evaluate on X_test and y_test
print("üéØ Voting Ensemble Accuracy:", voting.score(X_test, y_test))
```
